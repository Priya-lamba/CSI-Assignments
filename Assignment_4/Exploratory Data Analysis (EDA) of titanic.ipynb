```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
titanic.describe(include='all')
```

```python
missing = titanic.isnull().sum().sort_values(ascending=False)
```

```python
missing = missing[missing > 0]
```

```python
missing
```

```python
sns.heatmap(titanic.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Data Heatmap")
```

```python
titanic.hist(['age', 'fare'], bins=30, edgecolor='black')
plt.suptitle("Distribution of Age and Fare")
```

```python
sns.countplot(x='class', data=titanic)
plt.title("Passenger Class Distribution")
```

```python
print("\nSurvival Count:\n", titanic['survived'].value_counts())
sns.countplot(x='survived', data=titanic)
plt.title('Survival Count')
```

```python
sns.countplot(x='sex', data=titanic)
plt.title("Gender Distribution")
```

```python
print("\nAge Statistics:\n", titanic['age'].describe())
sns.histplot(x='age', data=titanic, kde=True)
plt.title('Age Distribution')
```

```python
sns.boxplot(x=titanic['age'])
plt.title("Boxplot of Age")
```

```python
sns.boxplot(x=titanic['fare'])
plt.title("Boxplot of Fare")
```

```python
sns.histplot(data=titanic, x='age', hue='survived', kde=True, multiple='stack')
plt.title("Age Distribution by Survival")
```

```python
sns.violinplot(x='survived', y='fare', data=titanic)
plt.title("Fare Distribution by Survival")
```

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Count by Gender")
```

```python
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
```

```python
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 65, 100], labels=['Child', 'Adult', 'Senior'])
print("\nSurvival by Age Group:\n", titanic.groupby('age_group')['survived'].value_counts())
sns.countplot(x='age_group', hue='survived', data=titanic)
plt.title('Survival by Age Group')
plt.show()
```

```python
print("\nSurvival by Gender and Class:\n", titanic.groupby(['sex', 'pclass'])['survived'].value_counts())
sns.catplot(x='sex', hue='survived', col='pclass', kind='count', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='embarked', hue='survived', data=titanic, palette='Spectral')
plt.title('Survival Based on Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic, x='fare', hue='survived', bins=20, kde=True, palette='viridis')
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Count')
```

```python
titanic['family_size'] = titanic['sibsp'] + titanic['parch']
plt.figure(figsize=(10, 6))
sns.countplot(x='family_size', hue='survived', data=titanic, palette='cubehelix')
plt.title('Survival Based on Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
```

```python
bins = [0, 12, 18, 40, 60, 80]
labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']
titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='survived', data=titanic, palette='magma')
plt.title('Survival Based on Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
```

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
sns.pairplot(titanic[['age', 'fare', 'survived']], hue='survived')
plt.suptitle("Pairwise Relationships")
```

```python
pd.crosstab(titanic['sex'], titanic['survived'], normalize='index').plot(kind='bar', stacked=True)
plt.title("Survival Rate by Gender")
```
